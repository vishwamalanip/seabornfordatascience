```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
import seaborn as sns
import pandas as pd
```

```python
y1=sns.load_dataset('penguins')


y1
```

```python
y1=sns.load_dataset('penguins')

sns.displot(y1['bill_length_mm'])

plt.show() 
```

```python
y1=sns.load_dataset('penguins')

sns.displot(y1['flipper_length_mm'])

plt.show() 
```

```python
y1=sns.load_dataset('penguins')

sns.displot(y1['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240]) #to see data properly organized use bins

plt.show() 
```

```python
y1=sns.load_dataset('penguins')

sns.displot(y1['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True) #karnel destiny estimator(kde) 

plt.show() 
```

```python
y1=sns.load_dataset('penguins')

sns.displot(y1['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True) #rug to see color density ,how it follows line

plt.show() 
```

```python
y1=sns.load_dataset('penguins')

sns.displot(y1['flipper_length_mm'],
            kde=True,rug=True,color='g',log_scale=True) #color='g' to change color of bar #log_Scale for to convert graph according tolog

plt.show() 
```
